```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Separate features and target
df_opelcorsa_combined = df_opelcorsa_combined.drop(columns=['ID','roadSurface', 'traffic'])

X = df_opelcorsa_combined.drop(columns=['drivingStyle'])
y = df_opelcorsa_combined['drivingStyle'].astype('category').cat.codes  # Convert labels to numeric

# Encode categorical features to numeric
label_encoder = LabelEncoder()
X_encoded = X.apply(label_encoder.fit_transform)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)

# Model 1: Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)

# Model 2: SVM
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)

# Evaluation Metrics
print("Random Forest Evaluation:")
print("Accuracy:", accuracy_score(y_test, rf_predictions))
print("Precision:", precision_score(y_test, rf_predictions, average='weighted'))
print("Recall:", recall_score(y_test, rf_predictions, average='weighted'))
print("F1 Score:", f1_score(y_test, rf_predictions, average='weighted'))
print("\nClassification Report:\n", classification_report(y_test, rf_predictions))

print("\nSVM Evaluation:")
print("Accuracy:", accuracy_score(y_test, svm_predictions))
print("Precision:", precision_score(y_test, svm_predictions, average='weighted'))
print("Recall:", recall_score(y_test, svm_predictions, average='weighted'))
print("F1 Score:", f1_score(y_test, svm_predictions, average='weighted'))
print("\nClassification Report:\n", classification_report(y_test, svm_predictions))
```

```python
# Random Forest
rf_train_accuracy = rf_model.score(X_train, y_train)
rf_test_accuracy = rf_model.score(X_test, y_test)

# SVM
svm_train_accuracy = svm_model.score(X_train, y_train)
svm_test_accuracy = svm_model.score(X_test, y_test)

print("Random Forest Training Accuracy:", rf_train_accuracy)
print("Random Forest Test Accuracy:", rf_test_accuracy)

print("SVM Training Accuracy:", svm_train_accuracy)
print("SVM Test Accuracy:", svm_test_accuracy)
```

```python
from sklearn.utils import resample



# Oversampling AggressiveStyle class
df_peugeot_even_pace = df_peugeot_combined[df_peugeot_combined['drivingStyle'] == 'EvenPaceStyle']
df_peugeot_aggressive = df_peugeot_combined[df_peugeot_combined['drivingStyle'] == 'AggressiveStyle']

df_opelcorsa_even_pace = df_opelcorsa_combined[df_opelcorsa_combined['drivingStyle'] == 'EvenPaceStyle']
df_opelcorsa_aggressive = df_opelcorsa_combined[df_opelcorsa_combined['drivingStyle'] == 'AggressiveStyle']

# Resample the minority class
df_opelcorsa_aggressive_upsampled = resample(df_opelcorsa_aggressive, replace=True, n_samples=len(df_opelcorsa_even_pace), random_state=42)
df_peugeot_aggressive_upsampled = resample(df_peugeot_aggressive, replace=True, n_samples=len(df_peugeot_even_pace), random_state=42)

# Combine the two classes
df_opelcorsa_balanced = pd.concat([df_opelcorsa_even_pace, df_opelcorsa_aggressive_upsampled])
df_peugeot_balanced = pd.concat([df_peugeot_even_pace, df_peugeot_aggressive_upsampled])
```

```python
df_opelcorsa_balanced['drivingStyle'] = df_opelcorsa_balanced['drivingStyle'].map({'EvenPaceStyle': 0, 'AggressiveStyle': 1})
df_peugeot_balanced['drivingStyle'] = df_peugeot_balanced['drivingStyle'].map({'EvenPaceStyle': 0, 'AggressiveStyle': 1})
```

```python
import numpy as np
# Drop non-numeric columns and the target column 'drivingStyle' for scaling
opelcorsa_features_to_scale = df_opelcorsa_balanced.select_dtypes(include=[np.number]).drop(columns=['drivingStyle'])
peugeot_features_to_scale = df_peugeot_balanced.select_dtypes(include=[np.number]).drop(columns=['drivingStyle'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
opelcorsa_scaled_features = scaler.fit_transform(opelcorsa_features_to_scale)
peugeot_scaled_features = scaler.fit_transform(peugeot_features_to_scale)
```

```python
# Convert the scaled features to a DataFrame
opelcorsa_scaled_df = pd.DataFrame(opelcorsa_scaled_features, columns=opelcorsa_features_to_scale.columns)
peugeot_scaled_df = pd.DataFrame(peugeot_scaled_features, columns=peugeot_features_to_scale.columns)

# Fill the NaN or NA containing rows with mean in both dataframes
opelcorsa_scaled_df.fillna(opelcorsa_scaled_df.mean(), inplace=True)
peugeot_scaled_df.fillna(peugeot_scaled_df.mean(), inplace=True)

# Add back the target and any other needed columns
opelcorsa_df_final = pd.concat([opelcorsa_scaled_df, df_opelcorsa_balanced[['drivingStyle']].reset_index(drop=True)], axis=1)
peugeot_df_final = pd.concat([peugeot_scaled_df, df_peugeot_balanced[['drivingStyle']].reset_index(drop=True)], axis=1)
```
